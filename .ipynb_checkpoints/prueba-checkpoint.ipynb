```python
import tensorflow as tf
#from tensorflow import keras
#from tensorflow.keras import layers
import time

# Verificar que la GPU est√° disponible y configurada
if tf.test.gpu_device_name():
    print('GPU encontrada')
else:
    print("No se encontr√≥ ninguna GPU")

# Crear un modelo de ejemplo
modelo = tf.keras.Sequential([
    tf.keras.layers.Dense(10, input_shape=(784,)),
    tf.keras.layers.Dense(10)
])

# Compilar el modelo
modelo.compile(optimizer='adam', loss='mse')

# Cargar los datos de ejemplo
(x_entrenamiento, y_entrenamiento), (x_prueba, y_prueba) = tf.keras.datasets.mnist.load_data()

# Preprocesar los datos
x_entrenamiento = x_entrenamiento.reshape(60000, 784).astype('float32') / 255.0
x_prueba = x_prueba.reshape(10000, 784).astype('float32') / 255.0

# Entrenar el modelo en la GPU
with tf.device('/device:GPU:0'):
    inicio = time.time()
    modelo.fit(x_entrenamiento, y_entrenamiento, epochs=100, batch_size=32)
    fin = time.time()

# Evaluar el modelo en la GPU
with tf.device('/device:GPU:0'):
    inicio_evaluacion = time.time()
    #loss, accuracy = modelo.evaluate(x_prueba, y_prueba)
    loss = modelo.evaluate (x_prueba, y_prueba, verbose=0)
    fin_evaluacion = time.time()

print(f'Tiempo de entrenamiento: {fin - inicio} segundos')
print(f'Tiempo de evaluaci√≥n: {fin_evaluacion - inicio_evaluacion} segundos')
#print(f'P√©rdida: {loss}, Precisi√≥n: {accuracy}')
print(f'P√©rdida: {loss}')
```
